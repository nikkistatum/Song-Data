```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import os
from sklearn import preprocessing

```

```python
df = pd.read_csv(os.path.join("Resources", "Spotify_Audio_Features.csv"))
df.head()
```

```python
#Encode artist id into quantitative
le = preprocessing.LabelEncoder()
df['track_id'] = le.fit_transform(df['track_id'].astype('str'))

#Encode artist id into quantitative
le = preprocessing.LabelEncoder()
df['track_name'] = le.fit_transform(df['track_name'].astype('str'))

#Encode artist id into quantitative
le = preprocessing.LabelEncoder()
df['artist_name'] = le.fit_transform(df['artist_name'].astype('str'))

```

```python
target = df["popularity"]
target_names = ["negative", "positive"]
```

```python
data = df.drop("popularity", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
# Loop through different k values to see which has the highest accuracy
# Note: We only use odd numbers because we don't want any ties
train_scores = []
test_scores = []
for k in range(1, 150, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 150, 2), train_scores, marker='o')
plt.plot(range(1, 150, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.show()
```

```python
# Note that k: 13 seems to be the best choice for this dataset
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print('k=1 Test Acc: %.3f' % knn.score(X_test, y_test))
```
