```python
from xgboost.sklearn import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import os
from sklearn import preprocessing
```

```python
df = pd.read_csv(os.path.join("Resources", "Spotify_Audio_Features.csv"))
df.head()
```

```python
drop_list = ['Unnamed: 0', 'artist_name', 'track_name', 'track_id', 'acousticness', 'duration_ms', 'instrumentalness', 'key', 'mode',
            'time_signature']
data = df.drop(drop_list, axis=1)
data.head()
```

```python
target = data["popularity"]
target_names = ["negative", "positive"]
```

```python
train = data.drop("popularity", axis=1)
feature_names = data.columns
train.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, target, random_state=42)
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
# make predictions for test data
y_pred = model.predict(X_test)
accuracy = model.score(X_test, y_test)
```
